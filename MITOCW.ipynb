```python
#Part A

total_cost = float(input("the cost of my dream home is: $ "))
annual_salary = float(input("each year I earn: $ "))
portion_saved = float(input("each month I will save: $ "))

monthly_salary = float(annual_salary/12)
months = 0

portion_down_payment = float(0.25*total_cost)
current_savings = 0.0 
total_saved = portion_saved*monthly_salary 
 

#how many months will it take to save enough money for downpayment?

while current_savings <= portion_down_payment:
    print(months)
    r = 0.04
    annual_return = (r/12)*current_savings
    months += 1
    current_savings += (total_saved + annual_return)
    
```
